# 3.1 График цен AMD

### Подключение необходимых библиотек

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

### Загружаем данных о котировках AMD

```python
AMD_info = pd.read_csv('AMD.csv', sep=',')
    
# Форматируем файл, удаляем лишние столбцы
cols = ['Open','High','Low','Adj Close','Volume']
AMD_info = AMD_info.drop(cols,axis=1)
```

```python
# Приводим дату к нужному нам формату
date = []

for i in range(len(AMD_info['Date'])):
    year = str(AMD_info['Date'][i])[:4]
    month = str(AMD_info['Date'][i])[5:7] 
    day = str(AMD_info['Date'][i])[8:]
    date.append(year + '-' + month + '-' + day) 

date = pd.to_datetime(AMD_info['Date'])
date = pd.DataFrame(date)

AMD_info['Date'] = date
```

```python
# Функция создания списка цен компании AMD за конкретный год
def amd_year_analysis(year):
    date = []
    price = []
    for i in range (len(AMD_info['Date'])):
        if AMD_info['Date'][i].year == year:
            date.append(AMD_info['Date'][i])
            price.append(AMD_info['Close'][i])
    return(date,price)
```

### Строим график цен AMD за 2016 и 2017 годы

```python
font = {'family': 'serif', 'serif': 'Times New Roman', 'weight': 'normal','size':14}
plt.rc('font',**font)

y_list = [2016,2017]
for year in y_list:
    s = 1
    fig,ax =plt.subplots(figsize=(9, 7))
    plt.ylabel('Цена, USD')
    year_list,price_list = amd_year_analysis(year)
    globals()['AMD_plot'+str(year)] = plt.plot(year_list,price_list, lw = 1.5, color = '#4dae7d')
    if year == 2016:
        plt.savefig('Рисунок 1. AMD: График цен за 2016 год.png', format = 'png')
    else:
        plt.savefig('Рисунок 2. AMD: График цен за 2017 год.png', format = 'png')
# Вывод рисунков: Рисунок 1. AMD: График цен за 2016 год, Рисунок 2. AMD: График цен за 2017 год
```

# Код для gif

```python
# AMD 2016
year = 2016
fig,ax =plt.subplots(figsize=(15, 11))
plt.tick_params(labelsize = 20)
ax.set_ylabel('Цена, USD')
year_list, price_list = amd_year_analysis(year)

# Создадим функцию анимации для нашего графика
def animate(i):
    ax.clear()
    ax.set_xlim([year_list[0],year_list[-1]])
    ax.set_ylim([1,13])
    ax.plot(year_list[0:i],price_list[0:i], lw = 2, color = '#4dae7d')
    ax.tick_params(axis='x', labelrotation = 45)
    ax.legend([f'Close, AMD {year}'], loc = 'upper left', fontsize = 'large') 
    return ax

iter_list_amd_2016 = []
for i in range(0, len(price_list)):
    iter_list_amd_2016.append(i)

# Вызываем функцию
amd_2016_animation = animation.FuncAnimation(fig, animate, frames = iter_list_amd_2016, repeat = False)

# Сохраняем интерактивный график в формате gif
# f = r"C:\Users\micha\YandexDisk\YandexDisk\УЧЕБНЫЕ МАТЕРИАЛЫ\II\МатСтат Курсовая\Моя Курсовая\Data\Беленький_ПМ20-4\AMD_2016.gif" 
f = r"C:\Users\micha\YandexDisk\УЧЕБНЫЕ МАТЕРИАЛЫ\II\МатСтат Курсовая\Моя Курсовая\Data\Беленький_ПМ20-4\AMD_2016.gif" 

writergif = animation.PillowWriter(fps=48) 
amd_2016_animation.save(f, writer=writergif)
```
