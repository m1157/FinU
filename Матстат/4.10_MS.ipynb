```python
from scipy.special import *
from scipy.stats import *
import numpy as np 
import math
import scipy as sp
import random
import matplotlib.pyplot as plt
import matplotlib.patches as pat
import matplotlib.ticker as ticker

font = {'family': 'serif', 'serif': 'Times New Roman', 'weight': 'normal','size':18}
plt.rc('font',**font)
```

```python
from collections import Counter
```

```python
iterations = 10000
n = 1000
```

```python
sample = []
for i in range(iterations):
    score = 0
    arif_mean = 0
    for value in range(n):
        score += random.randint(1,6)
    arif_mean = score/n
    sample.append(arif_mean)


    
x = np.linspace(3.32, 3.70, 100)
mu = 0
for i in range(iterations):
    mu += sample[i]
mu = mu/iterations
#  = 7/2 (1/6 * (1 + 2 + ... + 6))
#  = 1/6*(1 + 4 + ... + 36) - 49/4 = 35/12
sigma = np.sqrt(35/(12*n))
X=norm(loc=mu,scale=sigma)

fig,ax = plt.subplots(figsize=(13, 5))
plt.tick_params(labelsize = 16)
plt.tick_params(labelsize = 16)
plt.xlim(3.32, 3.68)
plt.grid(color='black', alpha=0.75, linestyle=':', linewidth=1)

plt.title(f'N={iterations}, n={n}', fontweight='normal')


plt.hist(sample, bins = 14, stacked = True, color = 'grey', edgecolor = 'k', alpha = 0.5, density = True)
plt.plot(x,X.pdf(x),color='red',lw=2)
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.02))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.8))


plt.show()
```
