```python
#1
n = 10
c = 2.23
X = t(n)

print(f'1. P = {X.cdf(c)}')
```

```python
X = t.pdf(x,10)
fig,ax =plt.subplots(figsize=(7, 7))
plt.tick_params(labelsize = 20)
plt.tick_params(labelsize = 20)
plt.grid(color='k', alpha=0.75, linestyle=':', linewidth=1)

plt.xlim(-3-0.1,3+0.1)
plt.ylim(0,0.5)

plt.title('1. PDF t(10)', fontweight='normal')
x = np.linspace(-3,3,2000)
plt.plot(x,X,color='c',lw=2)
ax.vlines(c, 0, 0.5, color = 'm', linestyle='--')
plt.legend([f't({n})',f'{c}'],loc = 'upper left')
plt.show()
```

```python
#2
n = 10
c_2 = -2.23
X = t(n)

print(f'2. P = {1 - X.cdf(c_2)}')
```

```python
X = t.pdf(x,10)
fig,ax =plt.subplots(figsize=(7, 7))
plt.tick_params(labelsize = 20)
plt.tick_params(labelsize = 20)
plt.grid(color='k', alpha=0.75, linestyle=':', linewidth=1)

plt.xlim(-3-0.1,3+0.1)
plt.ylim(0,0.5)

plt.title('2. PDF t(10)', fontweight='normal')
x = np.linspace(-3,3,2000)
plt.plot(x,X,color='c',lw=2)
ax.vlines(c_2, 0, 0.5, color = 'm', linestyle='--')
plt.legend([f't({n})',f'{c_2}'],loc = 'upper right')
plt.show()
```

```python
#3
n = 10
c_31 = 2.23
c_32 = - 2.23
X = t(n)

print(f'3. P = {X.cdf(c_31) - X.cdf(c_32)}')
```

```python
X = t.pdf(x,10)
fig,ax =plt.subplots(figsize=(7, 7))
plt.tick_params(labelsize = 20)
plt.tick_params(labelsize = 20)
plt.grid(color='k', alpha=0.75, linestyle=':', linewidth=1)

plt.xlim(-3-0.1,3+0.1)
plt.ylim(0,0.5)


plt.title('3. PDF t(10)', fontweight='normal')
x = np.linspace(-3,3,2000)
plt.plot(x,X,color='c',lw=2)
ax.vlines(c_31, 0, 0.4, color = 'm', linestyle='--')
ax.vlines(c_32, 0, 0.5, color = 'm', linestyle='--')
plt.legend([f't({n})',f'{c_31}',f'{c_32}'],loc = 'upper right')
plt.show()
```
