```python
from scipy.special import *
from scipy.stats import *
import numpy as np 
import math
import scipy as sp
```

```python
l = 10
r = 99

sample = []
for value in range(l, r):
    d = 2
    while value % d != 0:
        d += 1
    if d == value:
        sample.append(value)
```

###  

$$G(x_{1},x_{2},\ldots ,x_{n})={\sqrt[ {n}]{x_{1}x_{2}\cdots x_{n}}}=\left(\prod _{{i=1}}^{n}x_{i}\right)^{{1/n}}$$

```python
mlt_sample = 1
l = len(sample)
for i in range(l):
    mlt_sample *= sample[i]
G = (mlt_sample)**(1/l)
G
```

###  

$${\displaystyle H(x_{1},\ldots ,x_{n})={\frac {n}{{\frac {1}{x_{1}}}+{\frac {1}{x_{2}}}+\cdots +{\frac {1}{x_{n}}}}}={\frac {1}{{\frac {1}{n}}\sum \limits _{i=1}^{n}{\frac {1}{x_{i}}}}}}$$

```python
har_sum = 0
l = len(sample)
for i in range(l):
    har_sum += (1/sample[i])
H = l/har_sum
H
```

###   ( = 3)

$${\displaystyle A_{d}(x_{1},\ldots ,x_{n})={\sqrt[{d}]{\frac {\sum \limits _{i=1}^{n}x_{i}^{d}}{n}}}.}$$

```python
a = 3
power_sum = 0
l = len(sample)
for i in range(l):
    power_sum += sample[i]**a
A = (power_sum / l)**(1/3)
A
```
